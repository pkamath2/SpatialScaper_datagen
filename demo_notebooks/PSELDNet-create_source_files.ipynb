```python
for class_id in class_dict.keys():
    os.makedirs(Path(curated_dataset_dir)/class_id, exist_ok=True)
    class_data_arr = class_dict[class_id]
    for idx, class_data in enumerate(class_data_arr):
        aud, _ = librosa.load(class_data[0], sr=fsd_sr)
        aud = aud[int(class_data[1][0]*fsd_sr):int(class_data[1][1]*fsd_sr)]

        p = Path(curated_dataset_dir)/class_id/(str(Path(class_data[0]).stem)+'_'+str(idx)+'.wav')
        sf.write(file=p, data=aud, samplerate=fsd_sr)
    print('Completed', idx, class_id)
```
